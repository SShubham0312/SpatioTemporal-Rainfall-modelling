```python
import matplotlib.pyplot as plt
import numpy as np

# Define model names
models = ['XGBoost (SMOTE)','RandomForrest', 'GCN-Transformer', 'GCN-Transformer', '1D CNN']
# Sample metric values for each model (replace with actual values)
accuracy     = [0.75, 0.57, 0.7263, 0.7241, 0.414]
precision    = [0.56, 0.65, 0.4906, 0.4874, 0.01]
recall       = [0.69, 0.57, 0.8074, 0.7644,0.01]
f1_score     = [0.62, 0.61, 0.6103, 0.5952,0.01]
threat_score = [0.45, 0.44, 0.4392, 0.4237,0.01]
false_alarm  = [0.10, 0.05, 0.3029, 0.2904,0.01]

# Stack them for grouped bar chart
metrics = [accuracy, precision, recall, f1_score, threat_score, false_alarm]
metric_names = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'Threat Score', 'False Alarm Rate']

# Plot
x = np.arange(len(metric_names))
bar_width = 0.15
# ...existing code...
# Define a color palette for the bars
colors = ['#8E44AD','#2980B9', '#27AE60', '#F4D03F','#F39C12']  # Purple, Green, Yellow


plt.figure(figsize=(10, 6))
for i, model in enumerate(models):
    values = [metric[i] for metric in metrics]
    plt.bar(x + i * bar_width, values, width=bar_width, label=model, color=colors[i])

# Formatting
plt.xticks(x + bar_width, metric_names, rotation=45)
plt.ylim(0, 1.0)
plt.ylabel("Score")
plt.title("Model Performance Comparison")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
# ...existing code...
```
