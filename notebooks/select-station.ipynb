```python
import os
import shutil

# Source and destination directories
source_folder = "drive-download-20250702T130527Z-1-002/per_station_data/per_station_data"
destination_folder = "select-stations"

# Ensure destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# List of target place names
place_names = [
    "colaba", "santacruz", "panjim", "pernem", "dahanu", "jawhar", "mokheda",
    "palghar_agri", "vasai", "vikramgad", "wada", "bhira", "karjat_agri",
    "khalapur", "mahad", "mangaon", "matheran", "mhasla", "murud", "panvel_agri",
    "pen", "poladpur", "roha", "shriwardhan", "sudhagad_pali", "tala", "uran",
    "chiplun", "dapoli_agri", "guhagarh", "harnai", "lanja", "mandangad",
    "pawarwadi", "rajapur", "ratnagiri", "sangameshwar_devrukh", "savarde-arg",
    "devgad", "dodamarg", "kankavli", "kudal", "malvan", "mulde_agri",
    "sawantwadi", "vaibhavwadi", "vengurla arg", "canacona", "mormugao",
    "ssanguem", "ambernath", "bhiwandi", "kalyan", "murbad", "shahapur",
    "thane", "ulhasnagar"
]

# Normalize place names for matching
place_names_lower = [name.lower() for name in place_names]

# Go through all files in the source folder
for filename in os.listdir(source_folder):
    lower_filename = filename.lower()
    if any(place in lower_filename for place in place_names_lower):
        source_path = os.path.join(source_folder, filename)
        dest_path = os.path.join(destination_folder, filename)
        shutil.copy2(source_path, dest_path)
        print(f"Copied: {filename}")

print(f"Number of place names: {len(place_names)}")
print(f"Number of files copied: {len(os.listdir(destination_folder))}")
```
